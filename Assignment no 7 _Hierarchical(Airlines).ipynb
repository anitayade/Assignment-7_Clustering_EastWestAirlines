```python
air_data=data.drop(['ID#'],axis=1)
air_data
```

```python
#Normalization of the data
#Defining the normalization function
def normalf(i):
    x = (i - i.min())/(i.max()-i.min())
    return(x)
```

```python
#normalize the dataframe
air_data_normal = normalf(air_data.iloc[:,:])
air_data_normal
```

```python
# create dendrogram
plt.figure(figsize=(15, 6))  
dendograms=sch.dendrogram(sch.linkage(air_data_normal,'complete'))
```

```python
#Fit and Predict
#specifying hierarchical clustering, clusters as 5
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
#prediction - there is no centroid in agglomerative options
y_hc = hc.fit_predict(air_data_normal)
y_hc
```

```python
#5 clusters are formed.
#creating new dataframe column as clusters
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
data['h_clusterid'] = Clusters
```

```python
#No.of clusters formed
data['h_clusterid'].value_counts()
```

```python
data
```

```python
#grouping the customers as per clusters
data.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
#plotting the clusters based on the balance
plt.figure(figsize=(10, 7))  
plt.scatter(data['h_clusterid'],data['Balance'], c=hc.labels_)
```
